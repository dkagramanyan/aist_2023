```python
import glob, os, re
import pandas as pd
import matplotlib.pyplot as plt
from functools import reduce
import numpy as np
from transliterate import translit, get_available_language_codes
```

```python
translit_dict = {'а' : 'a', 'б' : 'b', 'в' : 'v', 'г' : 'g', 'д' : 'd', 'е' : 'ye', 'ё' : 'yo', 'ж' : 'zh', 'з' : 'z', 'и' : 'i', 'й' : 'j', 'к' : 'k', 'л' : 'l', 'м' : 'm', 'н' : 'n', 'о' : 'o', 'п' : 'p', 'р' : 'r', 'с' : 's', 'т' : 't', 'у' : 'u', 'ф' : 'f', 'х' : 'kh', 'ц' : 'ts', 'ч' : 'ch', 'ш' : 'sh', 'щ' : 'sch', 'ъ' : 'tzn', 'ы' : 'y', 'ь' : 'mzn', 'э' : 'e', 'ю' : 'yu', 'я' : 'ya'}
new_translit = dict((v,k) for k,v in translit_dict.items())
```

```python
files = glob.glob(r'E:\TD_OCR\part_1\**\*.jpg', recursive=True)
print(len(files))
```

```python
letter_dict = {}
for f in files:
    letter = f.split('\\')[-1].split('_')[0]
    if letter not in letter_dict:
        letter_dict[letter] = 1
    else:
        letter_dict[letter] +=1
        
        
letters_count = []
        
for k, v in letter_dict.items():
    if k.isalpha():
        rus = new_translit[k]
        letters_count.append([rus, v])
```

```python
letters_df = pd.DataFrame(letters_count, columns=['letter', 'count'])
letters_df.sort_values(by=['count'], ascending=False, inplace=True)
letters_df
```

```python
fig = plt.figure(figsize = (15, 10))
 
# creating the bar plot
plt.bar(letters_df['letter'].values, letters_df['count'].values, color ='maroon',
        width = 0.6)
plt.xlabel('Буквы', fontsize=15)
plt.ylabel('Количество образцов', fontsize=15)
plt.grid(axis='y')
plt.title("Распределение графем", fontsize=20)
plt.show()
```

```python
names = []

names_dirs = [x[0] for x in os.walk(r'E:\TD_OCR\part_1\names')]
for f in names_dirs:
    path = f.split('\\')
    if  len(path) == 5:
        string = path[-1].split('_')[0]
        work_id = int(path[-1].split('_')[1])
        names.append([work_id, string])
        
names_df = pd.DataFrame(names, columns = ['id', 'name'])
names_df.sort_values(by=['id'], ascending=True, inplace=True)
names_df.drop_duplicates(subset=['id'], keep=False, inplace=True)
names_df.reset_index(inplace=True, drop=True)
names_df
```

```python
surnames = []

surnames_dirs = [x[0] for x in os.walk(r'E:\TD_OCR\part_1\surnames')]
for f in surnames_dirs:
    path = f.split('\\')
    if  len(path) == 5:
        string = path[-1].split('_')[0]
        work_id = int(path[-1].split('_')[1])
        surnames.append([work_id, string])
        
surnames_df = pd.DataFrame(surnames, columns = ['id', 'surname'])
surnames_df.sort_values(by=['id'], ascending=True, inplace=True)
surnames_df.sort_values(by=['id'], ascending=True, inplace=True)
surnames_df.drop_duplicates(subset=['id'], keep=False, inplace=True)
surnames_df.reset_index(inplace=True, drop=True)
surnames_df
```

```python
codes = []

codes_dirs = [x[0] for x in os.walk(r'E:\TD_OCR\part_1\codewords')]
for f in codes_dirs:
    f = re.sub('.jpg', '', f)
    path = f.split('\\')
    if  len(path) == 5:
        string = path[-1].split('_')[0]
        work_id = int(path[-1].split('_')[1])
        codes.append([work_id, string])
        
codes_df = pd.DataFrame(codes, columns = ['id', 'codeword'])
codes_df.sort_values(by=['id'], ascending=True, inplace=True)
codes_df.sort_values(by=['id'], ascending=True, inplace=True)
codes_df.drop_duplicates(subset=['id'], keep=False, inplace=True)
codes_df.reset_index(inplace=True, drop=True)
codes_df
```

```python
data_frames = [names_df, surnames_df, codes_df]
combined_df = reduce(lambda  left,right: pd.merge(left,right,on=['id'], how='outer'), data_frames)
combined_df
```

```python
# В скольких случаях присутствуют все три элемента (name, surname, codeword):
combined_df['missing_vals'] = combined_df.isnull().sum(axis=1)
combined_df
```

```python
combined_df['missing_vals'].value_counts()
```

```python
rus_vals = []
cols = ['name', 'surname', 'codeword']


for i, row in combined_df.iterrows():
    translit_row = []
    for col in cols:
        text = row[col]
        if text not in [np.nan]:
            rus_text = translit(text, 'ru').lower()
            translit_row.append(rus_text)
        else:
            translit_row.append(np.nan)
    rus_vals.append(translit_row)          
```

```python
combined_df[['rus_name', 'rus_surname', 'rus_codeword']] = pd.DataFrame(rus_vals)
combined_df
```

```python
combined_df['merged'] = combined_df['rus_name'].fillna('') + combined_df['rus_surname'].fillna('') + combined_df['rus_codeword'].fillna('')
combined_df
```

```python
combined_df['char_count'] = combined_df['merged'].str.len()
combined_df.sort_values(by=['char_count'], ascending=False, inplace=True)
combined_df.reset_index(inplace=True, drop=True)
combined_df
```

```python
fig = plt.figure(figsize = (15, 10))
 
plt.scatter(combined_df.index, combined_df['char_count'].values, color ='maroon')
plt.grid(axis='y')
plt.xlabel('Количество образцов', fontsize=15)
plt.ylabel('Количество символов в образце', fontsize=15)
plt.title("Распределение авторов", fontsize=20)
plt.show()
```
